```python
df['Target'] = df['Target'].map({
    'Dropout':0,
    'Enrolled':1,
    'Graduate':2
})

df.corr()['Target']
```

```python
numerical_columns = ['Curricular units 1st sem (credited)', 'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)', 'Curricular units 1st sem (grade)', 'Curricular units 1st sem (without evaluations)', 'Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)', 'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (grade)', 'Curricular units 2nd sem (without evaluations)', 'Unemployment rate', 'Inflation rate', 'GDP', 'Age at enrollment']
categorical_columns = [col for col in df.columns if col not in numerical_columns and col != 'Target']

print(f"N.o. numerical columns: {len(numerical_columns)}")
print(f"N.o. categorical columns: {len(categorical_columns)}")
```

```python
df[numerical_columns].var()
```

```python
top_10_numeric = list(df[numerical_columns].var().index[0:10])

plt.figure(figsize=(10, 6))
correlation_matrix = df[top_10_numeric].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Top 10 Numerical Variables')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])
pca = PCA()
pca.fit(scaled_data)
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

pca = PCA()
pca.fit(scaled_data)

explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

explained_variance_df = pd.DataFrame({
    'Number of Principal Components': range(1, len(explained_variance) + 1),
    'Individual Explained Variance': explained_variance,
    'Cumulative Explained Variance': cumulative_explained_variance
})

plt.figure(figsize=(10, 6))
sns.lineplot(data=explained_variance_df, x='Number of Principal Components', y='Individual Explained Variance', marker='o', label='Individual Explained Variance')
sns.lineplot(data=explained_variance_df, x='Number of Principal Components', y='Cumulative Explained Variance', marker='o', label='Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Different Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import entropy

entropy_values = []
gini_values = []
for col in categorical_columns:
    counts = df[col].value_counts()
    ent = entropy(counts, base=2)
    entropy_values.append(ent)
    

variability_df = pd.DataFrame({
    'Categorical Variable': categorical_columns,
    'Entropy': entropy_values,
})

plt.figure(figsize=(7, 7))

sns.barplot(x='Entropy', y='Categorical Variable', data=variability_df.sort_values(by='Entropy', ascending=False), palette='viridis', hue='Categorical Variable', dodge=False)
plt.title('Entropy of Categorical Variables')
plt.xlabel('Entropy')
plt.ylabel('Categorical Variable')
plt.tight_layout()
plt.show()
```

```python
category_counts = {col: df[col].nunique() for col in categorical_columns}
category_counts_df = pd.DataFrame(list(category_counts.items()), columns=['Categorical Variable', 'Number of Categories'])

plt.figure(figsize=(7, 7))
sns.barplot(x='Number of Categories', y='Categorical Variable', data=category_counts_df.sort_values(by='Number of Categories', ascending=False), palette='viridis', hue='Categorical Variable')
plt.title('Number of Categories in Categorical Variables')
plt.xlabel('Categorical Variable')
plt.ylabel('Number of Categories')
plt.tight_layout()
plt.show()
```

```python
df[categorical_columns] = df[categorical_columns].astype(str)
df_lr = pd.get_dummies(df[categorical_columns], prefix_sep='#')
df_lr
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler


X = df_lr
y = df['Target']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
model = LogisticRegression(penalty='l1', solver='liblinear', class_weight='balanced')
model.fit(X, y)
coefficients = model.coef_[0]
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': coefficients})
feature_importance['Absolute_Importance'] = feature_importance['Importance'].abs()
feature_importance = feature_importance.sort_values(by='Absolute_Importance', ascending=False).head(20)

plt.figure(figsize=(7, 7))
sns.barplot(y=feature_importance['Feature'], x=feature_importance['Absolute_Importance'], palette='viridis', hue=feature_importance['Feature'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances from Logistic Regression with L1 Regularization')
plt.show()
```
